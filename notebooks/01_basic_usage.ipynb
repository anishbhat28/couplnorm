# 01 — Basic usage

Compare the coupling diagnostic **C** across distributions with the same
marginals but different joint structure. C sees the difference; a power
spectrum would not.

```python
import torch
from couplnorm import coupling_from_samples, CouplingMetric

torch.manual_seed(0)
```

## Synthetic generators

`coupled_field` multiplies every site of a sample by the same per-sample
scalar `z`, tying all Fourier amplitudes together and producing large
off-diagonal spectral-energy covariance.

```python
def independent_gaussian(B, N):
    return torch.randn(B, N)

def coupled_field(B, N, sigma=1.0):
    z = 1.0 + sigma * torch.randn(B, 1)
    eta = torch.randn(B, N)
    return z * eta
```

## C across distributions

```python
N = 32
datasets = {
    'independent':      independent_gaussian(10_000, N),
    'weakly coupled':   coupled_field(10_000, N, sigma=0.3),
    'strongly coupled': coupled_field(10_000, N, sigma=1.5),
}
for name, phi in datasets.items():
    C = coupling_from_samples(phi).item()
    print(f'{name:>20s}  C = {C:.4f}')
```

## The headline plot

```python
from couplnorm.plotting import plot_coupling_bars
import matplotlib.pyplot as plt

plot_coupling_bars(datasets)
plt.tight_layout()
plt.show()
```

Marginal Gaussianity does not imply joint independence. **C measures the gap.**
